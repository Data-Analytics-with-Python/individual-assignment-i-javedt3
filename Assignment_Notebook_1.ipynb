## 4. Add a new column (15pts)

* Add a new column **Customer_Days** to store the total number of days each customer has been with the company. You may follow the steps below:
  * use `pd.to_datetime(df["Dt_Customer"])` to convert the string variable to a *date-time* variable

  * use `today = datetime.today()` to find the date of today

  * use `(today - df["Dt_Customer"]).dt.days` to find the difference between today and the onbard date.

```python
from datetime import datetime

df["Dt_Customer"] = pd.to_datetime(df["Dt_Customer"], format='mixed', dayfirst=True)
today = datetime.today()
df["Customer_Days"] = (today - df["Dt_Customer"]).dt.days

display(df['Customer_Days'])
```

## 6. Numeric Variables (15pts)

* For `Income`, `Kidhome`, and `Teenhome`,
  * use boxplots to visualize their distributions.

* Based on these distributions, what can you say about the typical household profile of Nata's customers?

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(12,4))

plt.subplot(1,3,1)
df.boxplot(column="Income")
plt.title("Income")

plt.subplot(1,3,2)
df.boxplot(column="Kidhome")
plt.title("Kidhome")

plt.subplot(1,3,3)
df.boxplot(column="Teenhome")
plt.title("Teenhome")

plt.tight_layout()
plt.show()
```

Based on the boxplots, the typical household in Nata’s customer base has a moderate income level (most incomes are clustered in the lower to mid ranges, though a few customers are very high-income outliers). In terms of family composition, most households have 0 or 1 child and 0 or 1 teenager living at home, with very few families having 2 or more. This suggests that Nata’s customers are mainly small households with moderate incomes.

## 7. Spending by Products (15pts)

* Examine the spending amount columns `MntWines`,	`MntFruits`,	`MntMeatProducts`,	`MntFishProducts`,	`MntSweetProducts`,	`MntGoldProds`, and find the product with the highest total spending amount by all customers.



```python
spending_cols = ['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']
total_spending = df[spending_cols].sum()
print (total_spending)
```

```python
highest_product = total_spending.idxmax()
highest_value = total_spending.max()

print("Product with highest total spending:", highest_product)
print("Total spending amount:", highest_value)
```

After summing across all spending categories, Wine (MntWines) is the product with the highest total spending amount, at 676,083. This shows that wine purchases dominate overall customer spending, suggesting that wine is the most significant product category for Nata’s customers
